Vrať se k práci se souborem temperature.csv, který obsahuje informace o průměrné teplotě v různých městech v listopadu 2017.
 

    Vytvoř tabulku, která bude obsahovat údaje o teplotě za města Helsinki, Miami Beach a Tokyo.
    Vytvoř krabicový graf a porovnej rozsah teplot v těchto městech.


```python
import pandas

temperature = pandas.read_csv("temperature.csv")
temperature
```

```python
new_temperature = temperature[temperature["City"].isin(["Helsinki", "Miami Beach", "Tokyo"])]
new_temperature
```

```python
avg_temperature = new_temperature[["City", "AvgTemperature"]]
avg_temperature
```

```python
avg_helsinky = avg_temperature[avg_temperature["City"].isin(["Helsinki"])]
helsinky = avg_helsinky["AvgTemperature"]
helsinky
```

```python
avg_miami = avg_temperature[avg_temperature["City"].isin(["Miami Beach"])]
miami = avg_miami["AvgTemperature"]
miami
```

```python
avg_tokyo = avg_temperature[avg_temperature["City"].isin(["Tokyo"])]
tokyo = avg_tokyo["AvgTemperature"]
tokyo
```

```python
avg_box = helsinky.to_frame(name="Helsinky")
avg_box["Miami Beach"] = miami
avg_box["Tokyo"] = tokyo
avg_box
```

```python
avg_box.plot(kind="box")
```
